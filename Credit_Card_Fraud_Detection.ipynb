```python
# From the data set we are told that if the Class is equal to 0, then it is a legit transaction
# If the Class is equal to 1, then it is a fraudulent transaction
card_data_df['Class'].value_counts()
```

```python
fraud_transaction = card_data_df[card_data_df.Class == 1]
legit_transaction = card_data_df[card_data_df.Class == 0]
```

```python
# There are 492 fraudulent transactions and 284315 legit transactions
print(fraud_transaction.shape)
print(legit_transaction.shape)
```

```python
fraud_transaction.Amount.describe()
```

```python
legit_transaction.Amount.describe()
```

```python
card_data_df.groupby('Class').mean()
```

```python
# Now use Under-Sampling to build a smaple data set containing similar distrubtuion of fraudulent transactions and legit transactions
legit_sample = legit_transaction.sample(n = 492)
```

```python
new_dataset = pd.concat([legit_sample, fraud_transaction], axis=0)
```

```python
new_dataset.head()
```

```python
new_dataset.tail()
```

```python
new_dataset['Class'].value_counts()
```

```python
new_dataset.groupby('Class').mean()
```

```python
X = new_dataset.drop(columns='Class', axis = 1)
Y = new_dataset['Class']
```

```python
print(X)
```

```python
print(Y)
```

```python
#Split the data into test data and training data
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.2, stratify = Y, random_state = 20)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

**Classification using Logistic Regression**

```python
# Model training with Logistic Regression
logReg = LogisticRegression()
```

```python
logReg.fit(X_train, Y_train)
```

```python
X_train_predict = logReg.predict(X_train)
```

```python
training_data_acc = accuracy_score(X_train_predict, Y_train)
```

```python
print("Accuracy score: ", training_data_acc)
```

```python
X_test_predict = logReg.predict(X_test)
Test_data_acc = accuracy_score(X_test_predict, Y_test)
```

```python
print("Accuracy score: ", Test_data_acc)
```

```python
Y_pred = logReg.predict(X_test)
```

```python
accuracy_score(Y_test, Y_pred)
```
